```python
import numpy as np
import pandas as pd
import ast
import pickle
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv.zip')
movies
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv.zip')
credits
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies=movies.drop(['budget','homepage','original_language','original_title','popularity','production_countries','release_date','revenue','runtime',
             'spoken_languages','status','tagline','production_companies','movie_id'],axis=1)
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace= True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    c=0 #counter variable
    for i in ast.literal_eval(obj):
        if c!=3:
            c+=1
            L.append(i['name'])
        else:
            break
    return L 
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['keywords']+movies['cast']+movies['crew']+movies['genres']
```

```python
nmovies=movies[["id",'title','tags']]
```

```python
nmovies.head()
```

```python
nmovies['tags']=nmovies['tags'].apply(lambda x:" ".join(x))
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(stemmer.stem(i))
    return " ".join(y)
```

```python
nmovies['tags']=nmovies['tags'].apply(stem)
```

```python
nmovies.head()
```

```python
nmovies['tags']=nmovies['tags'].apply(lambda x:x.lower())
```

```python
nmovies.head()
```

```python
 from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(nmovies['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse= True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=nmovies[nmovies['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse= True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(nmovies.iloc[i[0]].title)
    
```

```python
recommend('American Pie')
```
